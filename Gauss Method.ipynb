Aplicando as Leis de Kirchoff no sistema:

Temos uma forma para $A\vec{x}=\vec{b}$: $$\begin{bmatrix}
1 & -1 & -1\\
R_1 + R_2 & 0 & R_4\\
0 & R_3 & -R_4
\end{bmatrix} \cdot \begin{bmatrix}
i_1\\ 
i_2\\
i_3\\
\end{bmatrix} = \begin{bmatrix}
0 \\
U_1-U_2 \\
U_2-U_3
\end{bmatrix} $$
Ou seja, substituindo os valores dados:
$$\begin{bmatrix}
1 & -1 & -1\\
11.9 & 0 & 1.8\\
0 & 5.3 & -1.8
\end{bmatrix} \cdot \begin{bmatrix}
i_1\\ 
i_2\\
i_3\\
\end{bmatrix} = \begin{bmatrix}
0 \\
15.0 \\
3.1
\end{bmatrix} $$
Que é equivalente ao proposto no enunciado.

```python
#Importando pacotes relevantes:
import numpy as np
```

```python
#Definindo as variáveis de entrada:
R1 = 4.7 #Ω
R2 = 7.2 #Ω
R3 = 5.3 #Ω
R4 = 1.8 #Ω
U1 = 24.0 #V
U2 = 9.0 #V
U3 = 5.9 #V
A = np.array([[0, R3, -R4], [R1+R2, 0, R4], [1, -1, -1]])
b = np.array([U2 - U3, U1 - U2, 0])
```

```python
#Construindo um solucionador do sistema de equações lineares:
def pivota(A,m):
    n = len(A)
    for i in range(m,n-1):
        if A[i][m] > A[m][m]:
            oldline = A[m]
            A[m] = A[i]
            A[i] = oldline
    return A
        
```
